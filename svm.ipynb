```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
en=OrdinalEncoder(categories=[['F','M']])
```

```python
df['Sex']=en.fit_transform(df[['Sex']])
```

```python
X=df.drop(columns=['Diagnosis (ALS)','ID'])
```

```python
y=df['Diagnosis (ALS)']
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_classif,SelectKBest,SequentialFeatureSelector
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
pca=PCA(n_components=40)
```

```python
sle=SelectKBest(f_classif,k=40)
```

```python
sfe=SequentialFeatureSelector(estimator=SVC(),n_features_to_select=40)
```

```python
model=SVC()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
pipe=Pipeline(steps=[
    ('scaler',scaler), 
    
    ('model',model)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
len(pipe.feature_names_in_)
```

```python
pipe.score(X_test,y_test)
```

```python
pipe1=Pipeline(steps=[
    ('scaler',scaler), 
    ('pca',pca),
    ('model',model)
])
```

```python
pipe1.fit(X_train,y_train)
```

```python
pipe1.score(X_test,y_test)
```

```python
param_grid = {
    'pca__n_components': [5, 10, 15, 20, 30, 40],  # PCA components
    'model__C': [0.1, 1, 10, 100],                   # SVC regularization
    'model__kernel': ['linear', 'rbf'],              # Kernel type
    'model__gamma': ['scale', 'auto', 0.01, 0.1, 1]  # Gamma for SVC
}
grid1=GridSearchCV(pipe1,param_grid,cv=5,scoring='accuracy')
```

```python
grid1.fit(X_test,y_test)
```

```python
grid1.score(X_test,y_test)
```

```python
grid1.best_params_
```

```python
y_pred=grid1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
